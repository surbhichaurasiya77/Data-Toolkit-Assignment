1. What is NumPy, and why is it widely used in Python?
- NumPy, short for Numerical Python, is a fundamental library for numerical and scientific computing in Python. It provides support for multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

 * NumPy is widely used for several reasons:-

    Efficiency :- NumPy arrays are implemented in C, making operations on them much faster than with standard Python lists, especially for large datasets.

    Functionality :- NumPy offers a wide range of mathematical, statistical, and linear algebra functions, making it suitable for various scientific and engineering applications.

    Integration :- NumPy seamlessly integrates with other popular Python libraries like Pandas, SciPy, and Matplotlib, forming the foundation of the scientific computing ecosystem in Python.

    Memory efficiency :- NumPy arrays store data in contiguous memory blocks, which is more memory-efficient than Python lists, particularly for large datasets.

    Broadcasting :- NumPy arrays support broadcasting, enabling operations between arrays of different shapes and sizes without explicit iteration.

6. What are the differences between NumPy arrays and Python lists?

    Mutability:- Elements are mutable.

17. Explain the difference between apply() and map() in Pandas.
- map() and apply() in Pandas are both used to apply a function to a Series or DataFrame, but they operate in different ways and are suited for different purposes.

 * map():-

   * Works on a Series element-wise.

   * It substitutes each value in a Series with another value, which can be determined by a function, a dictionary, or another Series.

   * It is optimized for element-wise transformations and is generally faster than apply() for this purpose.

   * Cannot be used on entire DataFrames directly.

   * It is suitable for simple one-to-one mappings or replacements.

 * apply():-

   * Can be used on both Series and DataFrames.

   * When used on a Series, it applies a function to each element, similar to map(), but it is more flexible as it can handle more complex operations.

   * When used on a DataFrame, it applies a function along an axis (rows or columns).

   * It can perform aggregations, transformations, and other operations that go beyond simple element-wise mappings.

   * It is more versatile than map() but can be slower for simple element-wise operations due to the overhead of function calls.

   * Can be used to apply a function to an entire row or column.

8. Create a line plot with multiple lines using Matplotlib.

9. A Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 32, 45, 28],
    'Score': [85, 90, 78, 88]
}
df = pd.DataFrame(data)

# Filter rows where Age > 30
filtered_df = df[df['Age'] > 30]

print(filtered_df)
```

```python
# Filter rows where Age > 30 and select only 'Name' and 'Score' columns
filtered_df = df.loc[df['Age'] > 30, ['Name', 'Score']]

print(filtered_df)
```

```python
# Filter using query method where Age > 30 and Score < 90
filtered_df = df.query('Age > 30 and Score < 90')

print(filtered_df)
```

```python
# Filter rows where Age > 30 and Score > 80
filtered_df = df[(df['Age'] > 30) & (df['Score'] > 80)]

print(filtered_df)
```

```python
# Filter rows where Age is either 25 or 45
filtered_df = df[df['Age'].isin([25, 45])]

print(filtered_df)
```

11. Perform matrix multiplication using NumPy.

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Method 1: Using the @ operator
result1 = A @ B

# Method 2: Using np.matmul()
result2 = np.matmul(A, B)

print("Result using @ operator:\n", result1)
print("\nResult using np.matmul():\n", result2)
```
